```python
import spacy#小于3.0 vector
nlp = spacy.load('en_core_web_lg')
```

```python
print(nlp.vocab['minister'].vector)
```

```python
print((nlp.vocab['today'].vector+nlp.vocab['is'].vector+nlp.vocab['a'].vector+nlp.vocab['good'].vector+nlp.vocab['today'].vector)/5)
print(nlp.vocab['today is a good day'].vector)
```

```python
import pandas as pd
sents = ["'Whitey on the Moon' is a 1970 spoken word"
         , "St Anselm's Church is a Roman Catholic church"
         , "Nymphargus grandisonae (common name: giant)"]

df=pd.DataFrame({"sentence":sents})

df['tokenized'] = df['sentence'].apply(nlp)
df['sent_vectors'] = df['tokenized'].apply(lambda x: x.vector)
# vectors = 0['sent_vector'].apply(pd.Series)
```

```python
print(df['sent_vectors'])
#3*300
```

```python
sents = ["'Whitey on the Moon' is a 1970 spoken word"
         , "St Anselm's Church is a Roman Catholic church"
         , "Nymphargus grandisonae (common name: giant)"]

tokenized=[nlp(i) for i in sents]
```
