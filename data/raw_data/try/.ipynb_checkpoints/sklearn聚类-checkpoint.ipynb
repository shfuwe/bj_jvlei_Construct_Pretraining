```python
import numpy as np
# 加载数据
def loadDataSet(fileName):
    data = np.loadtxt(fileName,delimiter='\t')
    return data
dataSet = loadDataSet("k-means实现test.txt")
k = 4
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

y_pred = KMeans(n_clusters=4, random_state=9).fit(dataSet)
print(y_pred)
print(y_pred.labels_)
print(y_pred.cluster_centers_)
print(y_pred.predict([[0, 0], [12, 3]]))
```

```python
plt.scatter(dataSet[:, 0], dataSet[:, 1], c=y_pred.labels_)
plt.scatter(y_pred.cluster_centers_[:,0], y_pred.cluster_centers_[:,1],c=[i for i in range(k)],marker='^')

plt.show()
```

```python
y_t=[1, 0, 2 ,3 ,1 ,0 ,2, 3 ,1, 0]
print(metrics.adjusted_rand_score(y_t, y_pred.labels_))
print(metrics.adjusted_mutual_info_score(y_t, y_pred.labels_)) 
```

```python
# dataSet = loadDataSet("test.txt")
import ast
def read_list2(text_path):
    lsit=[]
    with open('%s' % text_path, 'r', encoding="utf8") as f:  # 打开一个文件只读模式
        line = f.readlines()  # 读取文件中的每一行，放入line列表中
        for line_list in line:
            line = ast.literal_eval(line_list.replace('\n', ''))
            lsit.append(line)
    return lsit

modeled_datasets=read_list2('modeled_datasets.txt')
```

```python
print(len(modeled_datasets),len(modeled_datasets[0]))
```

```python
y_pred = KMeans(n_clusters=11, random_state=9).fit(modeled_datasets)
```

```python
def read_list(text_path):
    lsit=[]
    with open('%s' % text_path, 'r', encoding="utf8") as f:  # 打开一个文件只读模式
        line = f.readlines()  # 读取文件中的每一行，放入line列表中
        for line_list in line:
            lsit.append(line_list.replace('\n',''))
    return lsit
topic=read_list('./topic.txt')
```
