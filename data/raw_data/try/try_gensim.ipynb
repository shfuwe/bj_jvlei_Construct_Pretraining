```python
def read_list(text_path):
    lsit = []
    with open('%s' % text_path, 'r', encoding="utf8") as f:  # 打开一个文件只读模式
        line = f.readlines()  # 读取文件中的每一行，放入line列表中
        for line_list in line:
            lsit.append(line_list.replace('\n', ''))
    return lsit


def get_st():
    return read_list("../../st1.txt")
st = get_st()
data_list0 = read_list('./text.txt')
# data_list0=data_list0[0:1]
data_list=[]
i_n=0
for i in data_list0:
    if i_n%100==0:
        print(i_n)
    i_n=i_n+1
    for j in st:#去除标点符号
        i=i.replace(j,'')
    data_list.append(i)
# print(data_list)
```

```python
import sys
dl = []#每篇文章按词分开
i_n=0
for i in data_list:
    sys.stdout.write(str(i_n) + '\r')
    i_n=i_n+1
    l = i.split(' ')
    dl.append(l)
print(dl[0])
```

```python

def get_stop_words():
    # 加载停用词
    stop_words_path = "../../stopwords0.txt"
    stopwords = set([item.strip() for item in open(stop_words_path, 'r').readlines()])
    return stopwords
    # print(stopwords)


# 去掉文本中的停用词
def drop_stopwords(stopwords, line):
    line_clean = []
    for word in line:
        if word =='':
            continue
        if word in stopwords:
            continue
        line_clean.append(word)
    return line_clean

    
    
def clear(line):
    line_clean = []
    for word in line:
        if not word.isdigit():
            line_clean.append(word)
    return line_clean


clear_text = [clear(s) for s in dl]
print(clear_text[0])
stopwords = get_stop_words()
train_st_text = [drop_stopwords(stopwords, s) for s in clear_text]
print(train_st_text[0])
```

```python
import gensim.corpora as corpora
id2word = corpora.Dictionary(train_st_text)     # Create Dictionary
id2word.save_as_text("dictionary")                   # save dict
texts = train_st_text                           # Create Corpus
corpus = [id2word.doc2bow(text) for text in texts]   # Term Document Frequency
print(corpus[:1])
print([[(id2word[id], freq) for id, freq in cp] for cp in corpus[:1]])

#词袋模型
```

```python
from gensim import corpora, models, similarities
from pprint import pprint
def Tfidf(corpus):

    # initialize a model
    tfidf = models.TfidfModel(corpus)
    # print(tfidf)

    # Transforming vectors
    # 此时，tfidf被视为一个只读对象，可以用于将任何向量从旧表示（词频）转换为新表示（TfIdf实值权重）
#     doc_bow = [(0, 1), (1, 1)]
    # 使用模型tfidf，将doc_bow(由词,词频)表示转换成(词,tfidf)表示
    # print(tfidf[doc_bow])

    # 转换整个词库
    corpus_tfidf = tfidf[corpus]


    return corpus_tfidf

corpus_tfidf=Tfidf(corpus)

print(corpus_tfidf[:1])
print([[(id2word[id], freq) for id, freq in cp] for cp in corpus_tfidf[:1]])
```
